```python
import h5py
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
file_path = "/home/beams0/VNIKITIN/mona/azat/raven_pillar/raven_pillar_output.h5"
prefix_data='/exchange/data'
f = h5py.File(file_path, 'r')
```

```python
proj_id_0 = np.linspace(59,484,426).astype(int) #select projections from -85 to 85 degrees with 0.4 step
proj_id = np.delete(proj_id_0,316) #remove projection with mismatheched nscan and ndiff
ntheta = 426
nx = 121*2
ny = 154*2
theta = np.zeros(ntheta)
phase_map = np.zeros((ntheta,nx,ny))

for i in range(len(proj_id)):
    full_data_path = "{}/{}".format(prefix_data, proj_id[i])
    theta[i] = f[full_data_path].attrs['projection_angle']
    rot_corr = np.cos(np.deg2rad(theta[i]))
    scan = f[full_data_path+'/positions']
    data = np.int16(f[full_data_path+'/data'])
    grad_h = np.zeros(scan.shape[0])
    grad_v = np.zeros(scan.shape[0])
    for j in range(scan.shape[0]):
#         print(np.where(data[j] == np.amax(data[j])) #find center of diffraction pattern
        a = np.sum(data[j,0:data.shape[-2]//2,:])
        b = np.sum(data[j,data.shape[-1]//2::,:])
        c = np.sum(data[j,:,0:data.shape[-2]//2])
        d = np.sum(data[j,:,data.shape[-1]//2::])
        grad_h[j] = (a-b)/(a+b+1e-15)
        grad_v[j] = (c-d)/(c+d+1e-15)
    x = scan[:,0]
    y = scan[:,1]
    dx, dy = np.mgrid[np.min(x):np.max(x):nx*1j, np.min(y):np.max(y):ny*1j]
#     dx, dy = np.mgrid[0:nx:nx*1j, 0:ny:ny*1j]
    points = np.asarray([x,y]).T
    grid_h = griddata(points, grad_h, (dx, dy), method='linear',fill_value=np.mean(grad_h))
    grid_v = griddata(points, grad_v, (dx, dy), method='linear',fill_value=np.mean(grad_v))
    phase_map[i] = np.sqrt((grid_h-np.mean(grid_h))**2+(grid_v-np.mean(grid_v))**2)
    print('Done projection', i)
```
